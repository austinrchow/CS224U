```python
import pandas as pd
import numpy as np
import re
import csv
```

```python
data = pd.read_csv("text_emotion.csv")
data = data[data.sentiment != 'empty']
# data = data[data.sentiment != 'neutral']
data = data[data.sentiment != 'boredom']
# data = data[data.sentiment != 'boredom']

data = data[data.sentiment != 'surprise']
# print(surprise.content)
mapping = {'sadness': 0, 'worry':1, 'love': 2, 'hate': 3, 'anger':3,'enthusiasm': 4, 'fun': 4, 'happiness': 4, 'relief':4, 'neutral':5}
mappingR = { 0: 'sadness', 1: 'worry', 2: 'love', 3:'hate',4: 'happiness',5: 'neutral'}
sentiments = data["sentiment"].replace(mapping).tolist()
tweets = data["content"].tolist()
# for i in range(len(sentiments)):
    
# data.head()
```

```python
cleanedTweets = []
for tweet in tweets:
#     print(tweet)
    tweet =  tweet.lower()
    ats = re.findall(r"@\w+", tweet)
    for at in ats:
        tweet = tweet.replace(at, "")
    tweet = tweet.replace("&quot"," ")
    tweet = tweet.replace('&amp;', 'and')
    tweet = tweet.replace("..."," ")
    tweet = tweet.replace("!","")
    tweet = tweet.replace("?","")
    tweet = tweet.replace(".","")
    tweet = tweet.replace("\'","")
    tweet = tweet.replace("&lt;","")
    tweet = tweet.replace("&gt;","")
    tweet = tweet.strip()
    tweet = re.sub(' +',' ',tweet)
    tweet = ' '.join(re.sub("(@[A-Za-z0-9]+)|([^A-Za-z \t])|(\w+:\/\/\S+)", " ", tweet).split())
    cleanedTweets.append(tweet)
```

```python
wordsent = []
wordbody = []
for i,sentiment in enumerate(sentiments):
    print(sentiment, mappingR[sentiment])
    wordsent.append([i, mappingR[sentiment]])
    wordbody.append([i, cleanedTweets[i]])
wordsent = [["Id", "Tag"]] + wordsent       
wordbody = [["Id", "Body"]] + wordbody
# print(wordbody)
```

```python
with open('data/tweets_tags.csv', 'w') as myfile:
    wr = csv.writer(myfile)
    wr.writerows(r+[""] for r in wordsent)
    
with open('data/tweets.csv', 'w') as myfile:
    wr = csv.writer(myfile)
    wr.writerows(r + [""] for r in wordbody)
```
